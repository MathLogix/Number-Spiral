```python
def create_spiral(n):
    if n % 2 == 0:
        raise ValueError("The size of the spiral must be an odd number.")
    
    spiral = [[0] * n for _ in range(n)]
    
    x, y = n // 2, n // 2
    spiral[x][y] = 1
    
    directions = [(0, 1), (-1, 0), (0, -1), (1, 0)]
    current_direction = 0  
    steps = 1
    num = 2
    
    while num <= n * n:
        for _ in range(2):
            for _ in range(steps):
                x += directions[current_direction][0]
                y += directions[current_direction][1]
                if 0 <= x < n and 0 <= y < n:
                    spiral[x][y] = num
                    num += 1
            current_direction = (current_direction + 1) % 4
            
        steps += 1
    
    return spiral

def print_spiral(spiral):
    for row in spiral:
        print(" ".join(str(x).rjust(3, ' ') for x in row))

n = 11
spiral = create_spiral(n)
print_spiral(spiral)
```

```python
import turtle
import time

num_elements = 2048

def sieve_of_eratosthenes(n):
    prime = [True for _ in range(n + 1)]
    p = 2
    while p * p <= n:
        if prime[p]:
            for i in range(p * p, n + 1, p):
                prime[i] = False
        p += 1
    prime_numbers = [p for p in range(2, n + 1) if prime[p]]
    #if n >= 1:
    return prime_numbers

prime_numbers = sieve_of_eratosthenes(num_elements)

def draw_spiral(num_elements):
    screen = turtle.Screen()
    screen.setup(width=600, height=600)
    screen.bgcolor("black")
    screen.title("Spiral Animation")

    spiral_turtle = turtle.Turtle()
    spiral_turtle.speed(5)
    spiral_turtle.shape("circle")
    spiral_turtle.penup()
    spiral_turtle.color("white")

    spiral_turtle.hideturtle()

    x, y = 0, 0
    spiral_turtle.goto(x, y)
    spiral_turtle.pendown()
    spiral_turtle.dot(10)
    
    directions = [(10, 0), (0, 10), (-10, 0), (0, -10)]
    current_direction = 0
    steps = 1
    num = 1

    while num <= num_elements:
        for _ in range(2):
            for _ in range(steps):
                x += directions[current_direction][0]
                y += directions[current_direction][1]
                spiral_turtle.goto(x, y)               
                
                if num+1 in prime_numbers:
                    spiral_turtle.dot(10)

                num += 1
                #spiral_turtle.write(num, align="center", font=("Arial", 8, "normal"))
                                
                if num > num_elements:
                    break
                time.sleep(0)
            if num > num_elements:
                break
            current_direction = (current_direction + 1) % 4
        steps += 1

    screen.mainloop()

draw_spiral(num_elements-1)
```
